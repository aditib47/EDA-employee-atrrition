### Variable - BusinessTravel

```python
#uni-variate analysis
pd.crosstab(emp.BusinessTravel, columns="Count")
```

```python
# Bi-Variate
tab1=pd.crosstab(emp.Attrition, columns=emp.BusinessTravel)
tab1
```

```python
cototal=tab1.sum(axis=0)
cototal
rowtotal=tab1.sum(axis=1)
```

```python
#table of joint proportions

round(tab1/len(emp)*100,2)
```

```python
#Table of conditional proportion (Conditioned on rows)
(tab1.div(rowtotal, axis=0))*100
```

```python
#Table of conditional proportion (Conditioned on columns)

(tab1.div(cototal, axis=1))*100
```

```python
ax = sns.countplot(x="BusinessTravel",hue="Attrition", data=emp)
```

* out of Frequent travellers  the attrition is 24% highest compared to all categories
